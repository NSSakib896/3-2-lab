# convolution and correlation


```python
import numpy as np
import matplotlib.pyplot as plt
import scipy
```

```python
def function(x):
    if x>=1 and x<=3:
        if x>=1 and x<=2:
            return 2
        return 1
    else:
        return 0
```

```python
time = np.linspace(0, 5, 1000)
freq = 2
signal = 2*np.sin(2 * np.pi * freq * time)

impulse = [function(t) for t in time]
```

```python
plt.subplot(2, 1, 1)
plt.plot(time, signal)
plt.title('Signal')

plt.subplot(2, 1, 2)
plt.plot(time, impulse)
plt.title('Impulse')
plt.show()
```
