```python
import numpy as np
from scipy.signal import firwin,freqz
import matplotlib.pyplot as plt
```

```python
pass_band = 2000
stop = 5000
fs = 20000
filter_length = 21
```

```python
tap = firwin(filter_length, pass_band,width=stop-pass_band,fs=fs,window='hann')

response = freqz(tap, worN=8000)
```

```python
plt.plot(fs*response[0]/(2*np.pi), np.abs(response[1]))
plt.show()
```
