```python
import numpy as np
from scipy.signal import firwin, freqz
import matplotlib.pyplot as plt

# Filter specifications
fs = 20000  # Sampling frequency (20 kHz)
f_pass = 2000  # Passband edge frequency (2 kHz)
f_stop = 5000  # Stopband edge frequency (5 kHz)
N = 21  # Filter length

# Calculate normalized frequencies
f_pass_norm = f_pass / fs
f_stop_norm = f_stop / fs

# Design the filter using Hanning window
h = firwin(N, [f_pass_norm, f_stop_norm], pass_zero=False, window='hann')

# Frequency response
w, h_freqz = freqz(h, 1, worN=1000, fs=fs)

# Plot magnitude response
plt.figure()
plt.plot(w, 20 * np.log10(np.abs(h_freqz)))
plt.title("FIR Filter Magnitude Response")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Magnitude (dB)")
plt.grid(True)
plt.show()
```
